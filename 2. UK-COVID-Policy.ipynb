```python
import pandas as pd

# from url 'https://github.com/OxCGRT/covid-policy-tracker/blob/master/data/OxCGRT_latest.csv'
path = 'covid-policy-tracker/data/OxCGRT_latest.csv'
DF = pd.read_csv(path, low_memory=False)
DF
```

```python
UKDF = DF[DF['CountryName'] == 'United Kingdom'].reset_index(drop = True)
UKDF
```

```python

UK = UKDF[UKDF['RegionName'].isna()].reset_index(drop = True)
UK.to_csv('data/COVID_POLICY_UK.csv', index = False)
UK
```

```python
EnglandDF = UKDF[UKDF['RegionName'] == 'England'].reset_index(drop = True)
EnglandDF.to_csv('data/COVID_POLICY_ENGLAND.csv')
EnglandDF
```

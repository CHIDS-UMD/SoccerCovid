```python

start_date = pd.to_datetime('2019-06-01')
end_date = pd.to_datetime('2022-03-01')


All_Records = []
for idx, df in ltla_game_df.groupby('city'):
    
    df = df.sort_values('date').reset_index(drop = True)
    
    L = []
    for idx, row in df.iterrows():
        date = row['date']

        # last_game_date
        last_game_idx = idx - 1
        last_game_date = df.loc[last_game_idx]['date'] if last_game_idx > -1 else start_date

        # last_host_game_idx
        Flag = False
        for i in range(1, idx + 1):
            last_host_game_idx = idx - i
            if df.loc[last_host_game_idx]['host'] == True:
                Flag = True
                break
        if Flag == False: 
            last_host_game_idx = -1 
        last_host_game_date = df.loc[last_host_game_idx]['date'] if last_host_game_idx > -1 else start_date


        # next_game_date
        next_game_idx = idx + 1
        next_game_date = df.loc[next_game_idx]['date'] if next_game_idx < len(df) else end_date

        # next_host_game_idx
        Flag = False
        for i in range(1, len(df) - idx):
            next_host_game_idx = idx + i
            if df.loc[next_host_game_idx]['host'] == True:
                Flag = True
                break
        if Flag == False: 
            next_host_game_idx = len(df)
        next_host_game_date = df.loc[next_host_game_idx]['date'] if next_host_game_idx < len(df) else end_date

        row['last_game_date'] = last_game_date
        row['last_host_game_date'] = last_host_game_date
        row['next_game_date'] = next_game_date
        row['next_host_game_date'] = next_host_game_date

        L.append(row)


    new_df = pd.DataFrame(L)
    All_Records.append(new_df)
    
    

ltla_game_df = pd.concat(All_Records).reset_index(drop = True)

ltla_game_df
```

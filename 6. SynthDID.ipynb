# Game Cities vs Control Cities

```r
library(remotes)
library(synthdid)
path = 'data/soccer_covid_data.csv'

dataFocal = read.csv(path)

df = dataFocal[dataFocal$season == 'S1', ]
keeps<-c("areaName", "yweek_idx","new_case_per_10k" , "treat_post")
df<-df[keeps]
setup = panel.matrices(df)
tau.hat = synthdid_estimate(setup$Y, setup$N0, setup$T0)
se = sqrt(vcov(tau.hat, method='placebo'))
sprintf('point estimate: %1.2f', tau.hat)
sprintf('95%% CI (%1.2f, %1.2f)', tau.hat - 1.96 * se, tau.hat + 1.96 * se)
plot(tau.hat)


df = dataFocal[dataFocal$season == 'S2', ]
keeps<-c("areaName", "yweek_idx","new_case_per_10k" , "treat_post")
df<-df[keeps]
setup = panel.matrices(df)
tau.hat = synthdid_estimate(setup$Y, setup$N0, setup$T0)
se = sqrt(vcov(tau.hat, method='placebo'))
sprintf('point estimate: %1.2f', tau.hat)
sprintf('95%% CI (%1.2f, %1.2f)', tau.hat - 1.96 * se, tau.hat + 1.96 * se)
plot(tau.hat)

df = dataFocal[dataFocal$season == 'S3', ]
keeps<-c("areaName", "yweek_idx","new_case_per_10k" , "treat_post")
df<-df[keeps]
setup = panel.matrices(df)
tau.hat = synthdid_estimate(setup$Y, setup$N0, setup$T0)
se = sqrt(vcov(tau.hat, method='placebo'))
sprintf('point estimate: %1.2f', tau.hat)
sprintf('95%% CI (%1.2f, %1.2f)', tau.hat - 1.96 * se, tau.hat + 1.96 * se)
plot(tau.hat)
```

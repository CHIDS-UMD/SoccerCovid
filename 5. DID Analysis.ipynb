# Synth Control

```r
dataFocal = read.csv('data/gamecity/Birmingham.csv')
dataFocal = dataFocal[dataFocal$season == 'S1', ]
dataFocal
```

```r


library(remotes)

names(dataFocal)[names(dataFocal)=="Median.age"]<-"Median_age"
names(dataFocal)[names(dataFocal)=="People.per.sq..km"]<-"People_per_sq_km"
names(dataFocal)[names(dataFocal)=="Average.Score"]<-"Average_Score"
names(dataFocal)[names(dataFocal)=="Area..sq.km."]<-"Area_sq_km"


keeps<-c("areaName", 'areaidx', "yweek_idx","new_case_per_10k" ,
         "old_dependency_ratio" , "young_dependency_ratio",
         "popu_all","Median_age","People_per_sq_km",
         "Average_Score","Area_sq_km","male_avg_age",
         "female_avg_age","female_rate" 
         )

dataFocal<-dataFocal[keeps]
dataFocal
```

```r
require(tidysynth)
dataFocal%>% dplyr::glimpse()
```

```r

treat_city = dataFocal$areaName[1]
treat_time = 25
treat_time
treat_city
```

```r


dataFocal_out<-
  dataFocal %>%
  synthetic_control(outcome = new_case_per_10k,
                    unit=areaName,
                    time=yweek_idx,
                    i_unit=treat_city,
                    i_time=treat_time,
                    generate_placebos=F
  )%>%
  generate_predictor(time_window = 22:25, # time_window. The same. 
                     old_ratio = mean(old_dependency_ratio, na.rm = TRUE),
                     young_ratio=mean(young_dependency_ratio, na.rm= TRUE),
                     popu=mean(popu_all, na.rm= TRUE),
                     age=mean(Median_age, na.rm= TRUE),
                     peopleperkm=mean(People_per_sq_km, na.rm= TRUE),
                     Score=mean(Average_Score, na.rm= TRUE),
                     Areakm=mean(Area_sq_km, na.rm = TRUE),
                     maleage=mean(male_avg_age, na.rm= TRUE),
                     femaleage=mean(female_avg_age, na.rm= TRUE),
                     femalerate=mean(female_rate, na.rm = TRUE))%>%

  generate_predictor(time_window =19,
                     new_case_per_10k_19 = new_case_per_10k)%>%
  # generate_predictor(time_window =20,
  #                   new_case_per_10k_20 = new_case_per_10k)%>%
  generate_predictor(time_window =21,
                     new_case_per_10k_21 = new_case_per_10k)%>%
  generate_predictor(time_window =22,
                     new_case_per_10k_22=new_case_per_10k,)%>%
  generate_predictor(time_window =23,
                     new_case_per_10k_23=new_case_per_10k,)%>%
  generate_predictor(time_window =24,
                     new_case_per_10k_24=new_case_per_10k,)%>%
  generate_weights(optimization_window = 19:25,
                   margin_ipop = .02,sigf_ipop = 7,bound_ipop = 6)%>%

  generate_control()
```

```r
dataFocal_out %>% plot_trends()
```
